```python
import openpyxl
wb = openpyxl.load_workbook('testdata.xlsx')

sheet = wb.active 

transformed_data = []
current_record = []

for row in sheet.iter_rows(min_row=1, max_row=sheet.max_row, min_col=1, max_col=1):
    cell_value = row[0].value
    if cell_value is None:
        if current_record:
            transformed_data.append(' '.join(current_record))
            current_record  = []
    else:
        current_record.append(cell_value)

if current_record:
    transformed_data.append(' '.join(current_record))


for record in transformed_data:
    print(record)
```

```python
import openpyxl

# Load the Excel workbook
workbook = openpyxl.load_workbook('testdata.xlsx')

# Assuming the data is in the first sheet, specify the sheet name if needed
source_sheet = workbook.active

# Create a new workbook to store the transposed data
output_workbook = openpyxl.Workbook()
output_sheet = output_workbook.active

# Initialize variables to keep track of row and column positions in the output sheet
output_row = 1
output_column = 1

# Iterate through the rows in the source sheet
for row in source_sheet.iter_rows(values_only=True):
    for cell_value in row:
        # Skip blank cells
        if cell_value is not None and cell_value.strip() != "":
            # Write the non-blank cell value to the output sheet
            output_sheet.cell(row=output_row, column=output_column, value=cell_value)
            output_row += 1
        else:
            # Move to the next column in the output sheet when a blank cell is encountered
            output_row = 1
            output_column += 1

# Save the transposed data to a new Excel file
output_workbook.save('output.xlsx')

# Close both workbooks
workbook.close()
output_workbook.close()
```
