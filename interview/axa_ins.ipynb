```python
import os
import pandas as pd 
os.chdir(path='D:\Work\Interview_Questions\AXA insurance')
wb = ['policy_listing_jan_2013.xlsx','policy_listing_feb_2013.xlsx','policy_listing_mar_2013.xlsx','policy_listing_april_2013.xlsx',
'policy_listing_may_2013.xlsx']

sh = ['policy']
df_combined  = pd.DataFrame()

for file in wb:
    xls = pd.ExcelFile(file)
    for sheet in sh:
        df = pd.read_excel(xls, sheet_name=sheet)
        df_combined = df_combined.append(df,ignore_index=True)

df_combined.to_excel('policy_listing_policy_all_2013.xlsx', index=False)
```
