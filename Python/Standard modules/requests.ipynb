```python
import requests
from requests.exceptions import HTTPError
from bs4 import BeautifulSoup
url = "https://www.google.com"
url2 = "www.facebook.com"

try:
    response = requests.get(url)
    print(response)
    #print(response.text)
    print(BeautifulSoup(response.text))
except HTTPError as httperror:
    print(httperror)
except Exception as e:
    print(e)
else:
    print("Success")
```

```python
import requests

url = "https://httpbin.org/get"

r = requests.get(url)
print(r.status_code)
print(r.raise_for_status())


# GET requests
# Query string
# base_url + get/post/put + ?query string

payload = {"first_name":"John", "last_name" : "Hopkins"}

r = requests.get(url, params=payload)

# to get the url address
print(r.url)

# to view the contents - the response body in byte
print(r.content)

# to
print(r.text)


# POST requests
```
